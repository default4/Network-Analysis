```python
import networkx as nx
import community as community_louvain
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
```

```python
def load_data():
    G = nx.karate_club_graph()
    return G
```

```python
def community_detection(G):
    partition = community_louvain.best_partition(G)
    return partition
```

```python
def visualize_network(G, partition):
    pos = nx.spring_layout(G)
    cmap = mcolors.ListedColormap(list(mcolors.TABLEAU_COLORS)[:len(set(partition.values()))])

    nx.draw_networkx_nodes(G, pos, partition.keys(), node_size=200, cmap=cmap, node_color=list(partition.values()))
    nx.draw_networkx_edges(G, pos, alpha=0.5)
    nx.draw_networkx_labels(G, pos, labels={k: str(k) for k in G.nodes()}, font_size=12, font_color='black')

    plt.axis('off')
    plt.show()
```

```python
if __name__ == '__main__':
    G = load_data()
    partition = community_detection(G)
    visualize_network(G, partition)
```
